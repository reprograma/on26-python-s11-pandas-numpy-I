**Introdução**

Esse é o desafio do Titanic, que é o mais utilizado e indicado para pessoas que são iniciantes em ciência de dados. Nele é preciso prever se um passageiro sobreviveu ou não ao naufrágio, sabendo algumas características pessoais do mesmo. As bases de dados que serão utilizadas ao longo deste trabalho foram retiradas do desafio da rede Kaggle,2 que é uma das melhores plataformas para cientistas de dados. 

[Fonte](https://rpubs.com/modelthinkingbr/titanic-fgv)

Descrição de colunas:
* survival: variável binária que representa se o indivíduo sobreviveu (1) ou não (0);
* pclass: em qual classe ele estava alojado;
* name: nome do passageiro;
* sex: gênero do passageiro;
* age: idade do passageiro;
* sibsp: número de esposos e/ou irmãos à bordo;
* parch: número de pais ou filhos à bordo;
* ticket: número da passagem;
* fare: preço pago pela passagem;
* cabin: número da cabine em que estava alojado;
* embarked: cidade em que embarcou.

```python
# criando a função de processamento dos dados
def processamento_df(df):
    
    # 3 primeiras linhas do df
    df_3_linhas = df.head(3)
    
    # selecionando aleatoriamente 100 linhas do df
    df_menor = df.sample(100)
    
    # removendo colunas com muitos dados nulos
    df = df.drop(columns='Cabin')
    
    # ordenando por idade menor para o maior
    df = df.sort_values("Age")
      
    # removendo dados duplicados
    df = df.drop_duplicates()
    
    # reset index
    df = df.reset_index()
    
    # lista de colunas
    colunas = list(df.columns) 
    
    # metricas: analise descritiva
    metricas = df.describe(include='all')
    
    return df_3_linhas, df_menor, df, colunas, metricas
```

```python
# chamando a função processamento usando os dados do df
df_3_linhas, df_menor, df_teste, colunas, metricas = processamento_df(df)
```
