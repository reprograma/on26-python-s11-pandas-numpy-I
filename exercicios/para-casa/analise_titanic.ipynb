```python
import pandas as pd
df = pd.read_csv("C:\\Users\\nargy\\estudos\\Semana-11\\on26-python-s11-pandas-numpy-I\\material\\titanic.csv")
```

```python
print(df.isnull().sum())
```

```python
df = df.drop('Cabin', axis=1, errors='ignore') #excluindo a coluna cabin já que possui muitas lacunas
```

```python
df
```

```python
df = df.dropna(subset=['Embarque'])
```

```python
df = df.dropna(subset=['Idade'])
```

```python
#TRADUZINDO AS COLUNAS

#CRIA-SE UM DICIONÁRIO
traduzindo = {
    'PassengerId': 'Passageiro',
    'Survived': 'Sobrevivente',
    'Pclass': 'Classe',
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'SibSp': 'SibSp',
    'Parch': 'Pais&Filhos', #Parch = parents and children
    'Ticket': 'Bilhete',
    'Fare': 'Tarifa',
    'Embarked': 'Embarque'
}
```

```python
df = df.rename(columns=traduzindo)
```

```python
df
```

```python
#O gênero mais presente e a quantidade
genero_freq = df['Sexo'].mode()
genero_freq = genero_freq.to_list()
print(genero_freq)
```

```python
genero_freq_contagem= df['Sexo'].str.split().explode().value_counts()
N = 2
genero_freq = genero_freq_contagem.head(N)
print(genero_freq_contagem)
```

```python
idade_freq = df['Idade'].mode()
print(idade_freq)
```

É possível perceber que havia mais pessoas de gêneros masculinos do que pessoas de gênero feminino, sendo 453 masculinos e 259 femininos. A idade mais comum entre os passageiros é de 24 anos.
