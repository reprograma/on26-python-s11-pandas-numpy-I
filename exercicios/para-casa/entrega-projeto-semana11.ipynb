```python
# importanto a biblioteca pandas 
import pandas as pd
```

```python
# carregando os dados
df_filmes = pd.read_csv('IMDbMovies.csv')
df_filmes
```

```python
#fazendo um sort por diretor
df_filmes.sort_values(by=['Writer']).reset_index()
```

```python
#fazendo um sort por gêneros e tempo de duração
df_filmes.sort_values(by=['Main Genres', 'Runtime']).reset_index()
```

```python
df_filmes.info()
```

```python
# criando a função de processamento dos dados
def processamento_df(df_filmes):
    
    # 3 primeiras linhas do df
    df_3_linhas = df_filmes.head(3)
    
    # selecionando aleatoriamente 100 linhas do df
    df_menor = df_filmes.sample(100)
    
    # removendo colunas com muitos dados nulos
    df_filmes = df_filmes.drop(columns='Opening Weekend Gross in US & Canada')
    
    # ordenando por idade menor para o maior
    df_filmes = df_filmes.sort_values('Budget')
      
    # removendo dados duplicados
    df_filmes = df_filmes.drop_duplicates()
    
    # reset index
    df_filmes = df_filmes.reset_index()
    
    # lista de colunas
    colunas = list(df_filmes.columns) 
    
    # metricas: analise descritiva
    metricas = df_filmes.describe(include='all')
    
    return df_3_linhas, df_menor, df_filmes, colunas, metricas
```

```python
# chamando a função processamento usando os dados do df
df_3_linhas, df_menor, df_teste, colunas, metricas = processamento_df(df_filmes)
```

```python
# verificando se o retorno das 3 primeiras linhas está correto
df_3_linhas
```

```python
# verificando se o retorno dos dados aleatorios das 100 linhas está correto
df_menor
```

```python
# verificando se o retorno da lista de colunas
colunas
```

```python
# verificando se o retorno do tratamento dos dados está correto
df_teste
```
