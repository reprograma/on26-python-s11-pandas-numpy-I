```python
import pandas as pd
import numpy as np
```

# Base de dados

```python
df = pd.read_csv("titanic.csv")
df.head(5)
```

```python
# Exibição das informações gerais do DataFrame
print(df.info)
```

```python
# cópia de backup do DataFrame
df_backup = df.copy()
```

# Tratamento DataFrame

```python
#analisar colunas com os valorews nulos
print(df.isnull().sum())
```

```python
# Contagem de valores únicos em uma coluna ou no DataFrame
#print(df['Parch'].nunique)
print(df.nunique)
```

```python
#como a Cabine existems 687 valores nulos,usando o método drop() para excluir a coluna
df = df.drop('Cabin', axis=1)  # O parâmetro axis=1 indica que estamos excluindo uma coluna
```

```python
# Use o método drop_duplicates() para remover as linhas duplicadas
df = df.drop_duplicates()
```

```python
# Preencha os valores NaN na coluna 'age' com NaN
df['Age'].fillna(np.nan, inplace=True)
```

```python
df.head(10)
```

```python
# Resetar o index 
df = df.reset_index(drop=True)
```

```python
# Ordenação dos dados com base em uma coluna de idade dos mais novos para os mais velhos
df_sorted = df.sort_values(by='Age', ascending=True)
df_sorted
```

```python
# Vamos substituir os valores contidos na coluna 'Sex' por 'F' e 'M'
df['Sex'] = df['Sex'].replace({'male': 'M', 'female': 'F'})
```

```python
# Renomeie as colunas com nomes em português
df.rename(columns={
    'PassengerId': 'IdPassageiro',
    'Survived': 'Sobreviveu',
    'Pclass': 'Classe',
    'Name': 'Nome',
    'Sex': 'Sexo',
    'Age': 'Idade',
    'SibSp': 'SibSp',
    'Parch': 'Parch',
    'Ticket': 'Bilhete',
    'Fare': 'Tarifa',
    'Embarked': 'Embarque'
}, inplace=True)
```

```python
df.head(5)
```

# Insights

```python
# Insight 1: Média de idade dos passageiros
media_idade = df['Idade'].mean()
print(f'1. A média de idade dos passageiros é de {media_idade:.2f} anos.')
```

```python
# Insight 2: Quantidade de passageiros que sobreviveram (valor 1 na coluna 'Sobreviveu')
sobreviventes = df['Sobreviveu'].sum()
print(f'2. O número de passageiros que sobreviveram é {sobreviventes}.')
```

```python
# Insight 3: Contagem de passageiros por classe (Classe 1, Classe 2, Classe 3)
contagem_por_classe = df['Classe'].value_counts()
print(f'3. Contagem de passageiros por classe: {contagem_por_classe}')
```

# Exportar para o csv

```python
# Especifique o caminho do arquivo CSV para salvar o DataFrame
caminho_arquivo_csv = 'df_titanic.csv'
```

```python
# Exporte o DataFrame para um novo arquivo CSV
df.to_csv(caminho_arquivo_csv, index=False)  # O parâmetro index=False evita a gravação do índice no arquivo
```
