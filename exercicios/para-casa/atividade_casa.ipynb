### Introdução
A base de dados disponibilizada tem como base o incidente do Titanic, seus dados são compostos por uma ID para cada passageiro, status de sobrevivência em valor booleano (0 - morto e 1 - vivo), tipo de passagem, nome, sexo biológico, idade, ticket, valor pago, cabine e tipo de embarque.
Obs: os dados "SibSp" e "Parch" são foram possiveís de serem identificados.

### Processamento

```python
import pandas as pd
import numpy as np
import csv
```

```python
#Carregar dataframe
df = pd.read_csv('titanic.csv')
```

```python
#Ler as três primeiras linhas
df.head(3)
```

```python
#Ler a primeira linha (cabeçalho)
df.head(0)
```

```python
#Deleção de colunas
df = df.drop(columns=['SibSp', 'Parch'])
df
```

```python
#Sort de valores
df.sort_values(by='Survived')
```

```python
#Verifica se há valores nulos
print("Valores nulos por coluna do dataframe:")
print(df.isnull().sum())
```

```python
#Remove nulos em colunas específicas
df = df.dropna(subset=['Age', 'Cabin', 'Embarked'])
df.shape
```

```python
#Lista colunas
df.columns
```

```python
#Copia dataframe original
df.copy()
```

```python
#Reset index
df.reset_index(drop=True)
```

```python
#Tradução de colunas
df.rename(columns={'PassengerId': 'ID Passageiro', 'Survived': 'Sobrevivente', 'Pclass': 'Classe', 'Name': 'Nome', 'Sex': 'Sexo Biológico', 'Age': 'Idade', 'SibSp': 'Sib', 'Parch': 'Passe', 'Ticket': 'Passagem', 'Fare': 'Valor da Passagem', 'Cabin': 'Cabine', 'Embarked': "Embarque"}, inplace = True)
```

```python
#Exportar DF para csv
df.to_csv("Base_Titanic")
```
