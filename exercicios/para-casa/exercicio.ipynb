```python
# criando a funÃ§Ã£o de processamento dos dados
def processamento_df(df):
    
    # 3 primeiras linhas do df
    df_3_linhas = df.head(3)
    
    # selecionando aleatoriamente 100 linhas do df
    df_menor = df.sample(100)
    
    # removendo colunas com muitos dados nulos
    df = df.drop(columns='Cabin')
    
    # ordenando por idade menor para o maior
    df = df.sort_values("Age")
      
    # removendo dados duplicados
    df = df.drop_duplicates()
    
    # reset index
    df = df.reset_index()
    
    # lista de colunas
    colunas = list(df.columns) 
    
    # metricas: analise descritiva
    metricas = df.describe(include='all')
    
    return df_3_linhas, df_menor, df, colunas, metricas
```

```python
# chamando a funÃ§Ã£o processamento usando os dados do df
df_3_linhas, df_menor, df_teste, colunas, metricas = processamento_df(df)
```
