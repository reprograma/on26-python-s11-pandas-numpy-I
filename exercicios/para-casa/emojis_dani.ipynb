```python
import pandas as pd
import numpy as np

df = pd.read_csv("emojis.csv")
df

```

```python
#Mudando nome de Coluna
mapeamento_colunas = {
    'p HexId': 'Hex'
}

df.rename(columns=mapeamento_colunas, inplace=True)
df
```

```python
#Excluindo coluna
df.drop('Hex', axis=1, inplace=True)
df
```

```python
df.dtypes
#Mostrar tipo de dados na coluna
```

```python
df.columns
#Descri√ß√£o das colunas 
```

```python
#Mostrando valores nulos por colunas
print("Valores nulos por coluna do dataframe:")
print(df.isnull().sum())
```

```python
# Remove entradas (linhas) contendo valores nulos em colunas espec√≠ficas
df = df.dropna(subset=['Rank', 'Emoji','Year','Category','Subcategory','Name'])
df.shape
```

```python
df
```

```python
nome_da_coluna = 'Rank'

# Ordenando o DataFrame com base na coluna em ordem decrescente
df = df.sort_values(by=nome_da_coluna, ascending=False)
df
```

```python
nome_da_coluna = 'Rank'

# Ordenando a coluna especificada em ordem crescente
df = df.sort_values(by=nome_da_coluna)
df
```
