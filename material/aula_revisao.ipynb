```python
# normalização dos dados
dados_normalizacao = {'Altura': [1.65, 1.85, 1.88],
                      'Peso': [78, 91, 83],
                      'Porte': ['Pequeno', 'Grande', None]}

df_normalizacao = pd.DataFrame(dados_normalizacao)
df_normalizacao
```

```python
# O porte pode ser influenciado pelo peso (olhando o valor de 81.88 podemos classificar como pequeno)... O erro está pq não estamos normalizando os dados
# As duas variavies representam bem o Porte (Peso Altura), mas a grandeza das unidades (peso varia mais que a altura (varia em gradeza pouco entre as pessoas)). 
# A variavel peso tem um "peso" maior nessa análise, perante a variavel altura
df_normalizacao['A+P'] = df_normalizacao['Altura'] + df_normalizacao['Peso']
df_normalizacao
```

# Normalização Máx-Mín (escala Mín-Máx)



```python
# normalizando os dados
for column in df_normalizacao.columns:
    df_normalizacao["Altura_Normalizado"] = (df_normalizacao["Altura"] - df_normalizacao["Altura"].min()) / (df_normalizacao["Altura"].max() - df_normalizacao["Altura"].min())
    df_normalizacao["Peso_Normalizado"] = (df_normalizacao["Peso"] - df_normalizacao["Peso"].min()) / (df_normalizacao["Peso"].max() - df_normalizacao["Peso"].min())

df_normalizacao
```

```python
df_normalizacao['A+P_Normalizado'] = df_normalizacao['Altura_Normalizado'] + df_normalizacao['Peso_Normalizado']
df_normalizacao
```
