```python
# Obtendo a lista de gêneros únicos
generos_unicos = df['genres'].unique()
print("Gêneros Únicos:", generos_unicos)
```

```python
# Calcular a popularidade média por gênero
popularidade_por_genero = df.groupby('genres')['popularity'].mean().sort_values(ascending=False)
print("Popularidade Média por Gênero:")
print(popularidade_por_genero)
```
